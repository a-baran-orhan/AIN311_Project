```python
import numpy as np
import pandas as pd
import re
import string
```

```python
df = pd.read_csv("/content/udemy_ML.csv", header=None, sep=";")
```

```python
df
```

```python
df["outcomes"] = [row.split("{")[2] for row in df[0]]
```

```python
df
```

```python
df["outcomes"]=df["outcomes"].str.replace("'Outcomes':","")
df["outcomes"]=df["outcomes"].str.replace("[","")
df["outcomes"]=df["outcomes"].str.replace("]","")
df["outcomes"]=df["outcomes"].str.replace("}","")
df["outcomes"]=df["outcomes"].str.replace("\"","'")
df["outcomes"]=df["outcomes"].apply(lambda x : x + ',')
```

```python
df
```

```python
df.drop(columns=[0], inplace=True)
```

```python
df
```

```python
first_df = pd.DataFrame(columns=['text', 'course'])
pattern = "'(.*?)',"
for i, row in enumerate(df['outcomes']):
    lst = re.findall(pattern, row)
    temp_df = pd.DataFrame({'text': lst, 'course': [f'course {i}']*len(lst)})
    first_df = first_df.append(temp_df, ignore_index=True)
```

```python
first_df
```

```python
def cleaner (text):
    text = str(text).lower() # Lowering the case
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) # for punctation on text
    text = " ".join([word for word in text.split()])# cleaning the text to remove extra whitespace 
    
    return text
```

```python
first_df['text'] = first_df['text'].apply(cleaner)
```

```python
first_df
```

```python
final = pd.DataFrame(columns = ['word', 'course', 'sentence'])
for i, row in first_df.iterrows():
  course = row['course']
  lst = row['text'].split()
  temp_df = pd.DataFrame({'word': lst, 'course': [course]*len(lst), 'sentence': [f'Sentence {i}']*len(lst)})
  final = final.append(temp_df, ignore_index=True)
```

```python
final
```

```python
final.to_excel('ML_data.xlsx', encoding='utf-8', index=False)
```
